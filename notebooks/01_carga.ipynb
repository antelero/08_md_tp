# 01 - Cargas de datos

## Carga de datos de csv
Carga de fuentes principales, validación de estructura y aplicación de alcance del proyecto (departamentos en venta/alquiler).

## Salidas esperadas
- Dataset de alcance: outputs/interim/01_df_scope.csv
- Resumen de cargas: outputs/tablas/01_reporte_cargas.csv

### Parametros

```python
from pathlib import Path
import pandas as pd
#Parametros paths
out_interm = Path('outputs/interm')
out_tablas = Path('outputs/tablas')
pd.set_option('display.max_columns', 200)
pd.set_option('display.width', 220)
```

```python
df_venta = pd.read_csv('venta_departamentos_consolidado.csv')
df_alq = pd.read_csv('alquiler_departamentos_consolidado.csv')

df_venta['dataset_origen'] = 'venta'
df_alq['dataset_origen'] = 'alquiler'

df_total = pd.concat([df_venta, df_alq], ignore_index=True)

print('Filas venta:', len(df_venta))
print('Filas alquiler:', len(df_alq))
print('Filas total:', len(df_total))
print('Columnas:', df_total.shape[1])
```

```python
df_total.describe(include='all')
```

```python
# Definir alcance del proyecto: solo departamentos en venta/alquiler
f_type = df_total['type_building'].astype(str).str.strip().str.lower().eq('departamentos')
f_oper = df_total['type_operation'].astype(str).str.strip().str.lower().isin(['venta', 'alquiler'])
df_scope = df_total.loc[f_type & f_oper].copy()

reporte = pd.DataFrame({
    'metrica': ['filas_venta', 'filas_alquiler', 'filas_total', 'filas_scope', 'pct_scope'],
    'valor': [
        len(df_venta),
        len(df_alq),
        len(df_total),
        len(df_scope),
        round(100 * len(df_scope) / len(df_total), 2) if len(df_total) else 0,
    ]
})

df_scope.to_csv(out_interm / '01_df_scope.csv', index=False)
reporte.to_csv(out_tablas / '01_reporte_cargas.csv', index=False)

display(reporte)
print('Generado:', out_interm / '01_df_scope.csv')
print('Generado:', out_tablas / '01_reporte_cargas.csv')
```
